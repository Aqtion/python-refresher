# Simulation

```python
import numpy as np
```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 500  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = 0.01  # s
time_max = 10  # s
```

```python
time = 0  # s
depth = 5  # m
velocity = 0 # m/s
acceleration = 0  # m/s^2
buoyancy =  density_water * object_volume * g # N
weight = density_object * object_volume * g  # N
```

```python
# lists for plotting later
time = np.arange(0, time_max, time_step)
position = np.zeros_like(time)
velocity = np.zeros_like(time)
acceleration = np.zeros_like(time)
```

```python
# simulation loop
for i in range(1, len(time)):
    # calculate the force on the object
    print(f"Force:{buoyancy - weight}")
    # calculate acceleration
    acceleration[i] = (buoyancy-weight)/(density_object*object_volume)
    print(f"Acceleration:{acceleration[i]}")
    # calculate velocity
    velocity[i] = time_step * acceleration[i] * i
    print(f"Velocity:{velocity[i]}")
    # calculate position
    position[i] = 0.5 * acceleration[i] * pow(i * time_step, 2)
    print(f"Position:{position[i]}")
    
    
```

# Plot the results

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(time, position, label="Position")
plt.plot(time, velocity, label="Velocity")
plt.plot(time, acceleration, label="Acceleration")
plt.xlabel("Time (s)")
plt.ylabel("Position (m), Velocity (m/s), Acceleration (m/s^2)")
plt.legend()
plt.show()
```
